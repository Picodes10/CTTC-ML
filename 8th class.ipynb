```python
plt.bar(iris.sepal_length,iris.petal_length,color='cyan',label='col_cyan',width=0.1)
plt.bar(iris.sepal_width,iris.petal_width,color='grey',label='col_grey',width=0.1)
plt.legend()
plt.show()
```

```python
plt.bar(iris.petal_length[0:10],iris.petal_width[0:10],color='pink',width=0.1)
plt.bar(iris.sepal_length[0:10],iris.sepal_width[0:10],color='yellow',width=0.1)
plt.show()
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width,marker='H',color='yellow',s=20)
plt.scatter(iris.petal_length,iris.petal_width,marker='H',color='orange',s=20)
plt.grid(True)
plt.show()
```

# seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv(r"C:\Users\DELL\Downloads\iris\iris.data",header=None)
iris
```

```python
iris.columns = ['sl','sw','pl','pw','Species']
iris
```

```python
# kernel density estimation graph(kde)
```

```python
sns.kdeplot(iris.sl)
plt.show()
```

```python
# displot (histogram, baragraph)
```

```python
sns.displot(iris.sl)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# dist plot (histogram, kde)
```

```python
sns.distplot(iris.sl)
plt.show()
```

```python
# count plot (it can show the frequency o fthe object column)
```

```python
sns.countplot(x=iris.Species)
plt.show()
```

```python
# boxplot (1st numeric_column, 2nd object column)
```

```python
sns.boxplot(y=iris.sl,x=iris.Species)
plt.show()
```

```python
sns.violinplot(y=iris.sl,x=iris.Species)
plt.show()
```

```python
sns.boxenplot(y=iris.sl,x=iris.Species)
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
# Heat Map
```

```python
iris1 = iris.drop(['Species'],axis=1)
iris1
```

```python
cor = iris1.corr()
cor
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(cor,annot=True)
plt.show()
```
